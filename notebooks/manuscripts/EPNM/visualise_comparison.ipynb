```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load data
df = pd.read_csv('comparison_accuracy.csv', index_col=[0,1,2]).squeeze()
# Subtract minimum
df -= df.loc['Alleman et al.', 'Alleman et al.', 'Alleman et al.']
```

```python
fig,ax=plt.subplots(nrows=1, ncols=2, sharex=False, sharey=False, figsize=(8.3,11.7/2.5))

# left
ax[0] = sns.heatmap(np.transpose(df.loc[slice(None), 'Pichler et al.', slice(None)].values.reshape([2,2])),
                    xticklabels=['del Rio-Chanona et  al.','ERMG survey'],
                    yticklabels=['Pichler et al.','Alleman et al.'],
                    annot=True, fmt=".2f", vmin=0, vmax=1.5, cbar=False, ax=ax[0], square=True,
                    annot_kws={"fontsize": 11},cmap="Blues")
ax[0].set_xlabel('Labor supply', size=11)
ax[0].set_ylabel('Other demand', size=11)
ax[0].tick_params(labelsize=11)
ax[0].set_title('Household demand:\nCongressional Budget Office', size=11)

# right
ax[1] = sns.heatmap(df.loc[slice(None), 'Alleman et al.', slice(None)].values.reshape([2,2]),
                    xticklabels=['del Rio-Chanona et  al.','ERMG survey'],
                    yticklabels=[],
                    annot=True, fmt=".2f", vmin=0, vmax=1.5, cbar=False, ax=ax[1], square=True,
                    annot_kws={"fontsize": 11},cmap="Blues")
ax[1].set_title('Household demand:\nCalibrated (Alleman et al.)',size=11)
ax[1].tick_params(labelsize=11)
ax[1].set_xlabel('Labor supply', size=11)

fig.tight_layout()
plt.tight_layout()
plt.savefig('comparing_shocks.pdf')
plt.show()
plt.close()
```
